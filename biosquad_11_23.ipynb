## Custom Function for importing LCC XML files
We are using the xml.etree.ElementTree library to grab and process our data

```python
# create a function for importing all XML LCC data files from GitHub repositroy
def parse_lcc_xml(url):
    response = requests.get(url)
    if response.status_code != 200:
        print(f"Failed to download {url}: {response.status_code}")
        return None
    root = ET.fromstring(response.content) 

    # extract the feature names
    feature_names = [feature.find('Name').text for feature in root.findall('.//GDALRasterAttributeTable/FieldDefn')]

    # extract the rows
    rows = []
    for row in root.findall('.//GDALRasterAttributeTable/Row'):
        rows.append([f.text for f in row.findall('F')])

    lcc_df = pd.DataFrame(rows, columns=feature_names)

    # convert to values in each column to numeric values and pass if values are strings
    for col in lcc_df.columns:
        try:
            lcc_df[col] = pd.to_numeric(lcc_df[col])
        except ValueError:
            pass

    return lcc_df
```

### Converting Pixels to Area (square kilometer)

Land cover changes are visualized in pixels on an image file, or *.tif*. Here, we can visualize spatial data by converting pixel counts for each main LCC group into an area. We are in the process of writing a mapping code for to take each $900m^2$ pixel and turn it into $km^2$

```python
# convert pixels to km^2
def pixels_to_km2(pixel_count):
    return pixel_count * 0.0009      #area in km^2 of one 30x30 pixel

trim_lcc_df['area'] = pixels_to_km2(trim_lcc_df['Pixel Count'])

trim_lcc_df
```

## Bring All Linear Regressions Together in One Figure: Individual CO2 ~ Area of Land Conversion for Each Category
To further test for and plot linear regressions between the LCC groups and CO2 emissions, we first put the separate dataframes for each type of land conversion into one data dictionary for clean looping when plotting. We also standardized the scales in each regression line to enable more precise visual comparison of the amount of land area changed within each examined category. 

First, we plot all the linear regressions (using `scikit-learn`) of the annual increase of CO2 emissions to LCC area from 2000 to 2020 for every LCC group. Annual CO2 increase (ppm) is on the x-axis and the annual area of land conversion (in kilometers squared)  on the y-axis. The slope and R2 of each linear regression are included in the legend. LCC main groups are distinguished by color, and the linear regression fit lines are included as solid lines. 

```python
# make every year value be an integer for standarization between datasets
for df in [FtD, ItO, VtB, FtA, WtI]:
    df["year"] = df["year"].astype(int)
CO2_trim["year"] = CO2_trim["year"].astype(int)

# put all groups into one single dictionary for clean looping while plotting later
groups = {
    "Forest → Developed": FtD,
    "Ice → Ocean": ItO,
    "Vegetation → Barren": VtB,
    "Forest → Agriculture": FtA,
    "Wetlands → Impervious": WtI
}

# make colors list to color code figure
colors = ['blue', 'green', 'red', 'purple', 'orange']

# figure dimensions
plt.figure(figsize=(14, 6))

for (label, df), color in zip(groups.items(), colors):

    # merge land area and CO₂ data, by year
    merged = df.merge(CO2_trim[['year', 'ann inc']], on="year", how="inner")

    X = merged['ann inc'].values.reshape(-1, 1)
    y = merged['area'].values

    # Fit regression **corrected line**
    model = LinearRegression()
    model.fit(X, y)
    y_pred = model.predict(X)

    # get stats
    slope = model.coef_[0]
    r2 = r2_score(y, y_pred)

    # sort for clean regression line
    sort_idx = X[:, 0].argsort()
    X_sorted = X[sort_idx]
    y_pred_sorted = y_pred[sort_idx]

    # plot points and the regression lines
    plt.scatter(X, y, color=color, alpha=0.6, label=f"{label} data")
    plt.plot(X_sorted, y_pred_sorted, linestyle='-', color=color,
             linewidth=2, label=f"{label} trend (slope={slope:.3f}, R²={r2:.2f})")

   

plt.xlabel("CO₂ (ppm/year) Increase", fontsize=12)
plt.ylabel("Area (km$\mathregular{^2}$)", fontsize=12)

plt.grid(alpha=0.3)
plt.tight_layout()

plt.savefig("areavsCO2.png", dpi=300, bbox_inches='tight')

plt.show()

# remove the outlier in vegetation --> barren category????
```

## Linear Regressions against LCC Timeseries Data
Next, we apply the linear regression for each group over trends in LCC area changes from 2000 to 2020 in a single plot. Year is on the x-axis and area (square kilometers) is on the y-axis. LCC groups are distinguished by color. Trends in area overtime are plotted as points connected by a solid line, and the linear regression fit lines are included as dashed lines. 

```python
## use the same colors list and dictionary of land change dfs for this figure

# figure dimensions
plt.figure(figsize=(10, 6))

for (label, df), color in zip(groups.items(), colors):

    # extract X and y from each df
    X = df['year'].values.reshape(-1, 1)
    y = df['area'].values

    # fit linear regression modles
    model = LinearRegression()
    model.fit(X, y)
    y_pred = model.predict(X)

    X_sm = sm.add_constant(X)
    sm_model = sm.OLS(y, X_sm).fit()
    p_value = sm_model.pvalues[1]
    
    slope = model.coef_[0]
    r2 = r2_score(y, y_pred)

    print(label, "p-value:", p_value)
    # plot actual data in the figure
    plt.plot(df['year'], df['area'], marker='o', color=color, label=label)

    # now code regression line
    plt.plot(df['year'], y_pred, linestyle='--', color=color, alpha=0.7, label=f"{label} trend (slope={slope:.3f}, R²={r2:.2f})")

# plot labs and legend
plt.xlabel("Year", fontsize=12)
plt.ylabel("Area (km$\mathregular{^2}$)", fontsize=12)
plt.title("Land Cover Change Trends with Linear Regressions", fontsize=14)
plt.legend()

# formatting
plt.grid(alpha=0.3)
plt.tight_layout()
plt.xticks(np.arange(2000, 2021, 2))
plt.show()
```
